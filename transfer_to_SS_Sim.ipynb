```python
# import the image "1d echo.png" and display it
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

img = Image.open("1d echo.png")
amplitude = np.array(img)
plt.imshow(amplitude, cmap='Greys', aspect='auto')
plt.gca().invert_yaxis()
plt.show()
print((amplitude.shape))
print("height, width, channels(?)")
```

```python
# plot the value of summed channels vs the index
y_start = 18  # nanoseconds
y_end = 32  # nanoseconds
plt.figure(figsize=(2, 4))
f, (a0, a1) = plt.subplots(1, 2, gridspec_kw={'width_ratios': [4, .5]})
f.set_tight_layout(True)

plt.yticks(np.linspace(0, 1276, 5), np.linspace(y_start, y_end, 5))
# plot the 1d image as a  waveform in the second subplot with the x axis labeled from 18 to 32 
y = np.linspace(y_start, y_end, 1276)
x = np.sum((255 - amplitude), axis=1)
a0.plot(x, y, 'black', linewidth=0.5)
# a0.invert_yaxis()

a0.set_title("Amp vs Time")
a0.set_xlabel("Amplitude")
a0.set_ylabel("Time (ns)")
# make the y axis go from 18 to 32
plt.ylim(y_start, y_end)
# invert the x axis
f.set_size_inches(4, 8)
plt.gca().invert_xaxis()

# plot the image in the first subplot with the y axis labeled from 18 to 32 nanoseconds
a1.imshow(amplitude, cmap='Greys', aspect='auto', extent=[0, 1, y_start, y_end])
# invert the y axis
a1.invert_yaxis()
# set the title of the plot
a1.set_title("1D Image")
# don't show x axis labels or ticks
a1.set_xticks([])


# plt.show()
f.savefig("1d_echo.png", dpi=300)
```
